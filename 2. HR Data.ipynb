```python
def salary(x):
    if x == 'low':
        return 1
    elif x == 'medium':
        return 2
    else:
        return 3
```

```python
def satisfaction(x):
    if x < .12:
        return 1
    elif x < .31:
        return 2
    elif x < .47:
        return 3
    else:
        return 4
```

```python
df['New_Sat'] = df['satisfaction_level'].apply(lambda x: satisfaction(x))
```

```python
def satisfactionn(x):
    if x < .12:
        return 1
    elif x < .31:
        return 2
    elif x < .47:
        return 1
    else:
        return 2
```

```python
df['NewGroup_Sat'] = df['satisfaction_level'].apply(lambda x: satisfactionn(x))
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
X = df[['satisfaction_level', 'last_evaluation', 'number_project',
       'average_montly_hours', 'time_spend_company', 'Work_accident',
       'promotion_last_5years', 'New_Sal', 'New_Sat',
       'NewGroup_Sat']]
y = df['left']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
